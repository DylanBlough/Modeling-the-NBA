## "This is How I Win:" Predicting the Outcomes of NBA Games



Using team stats from the first half of the NBA season, I am going to use Linear Regression and KNearest Neighbors to predict the winner of regular season and playoff NBA games over the course of the second half of the 2019-2020 season. I will base my analysis on Dean Oliver's 4 Factors philosophy: https://squared2020.com/2017/09/05/introduction-to-olivers-four-factors/

Once I have decided on strongly correlated factors to winning games, I will use data from sports gambling websites to weight my predicted outcomes to create a hypothetically profitable sports betting model. I plan on validating the model using last year's completed NBA seasons as a training/test set. My metric for success will be correctly predicting 55% of the outcomes of games.

### Data Sources

Team statistics: BasketballReference.com
Historic betting statistics: OddsSharks.com (to be included)

```python
# Imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in the dataset

df = pd.read_csv('./2020_team_stats.csv', index_col = None)
```

## Data Cleaning & EDA

```python
df.head()
```

```python
df.columns
```

```python
# check the datatypes out

df.dtypes
```

```python
# we don't need the age, arena, or attendance metrics for predicting games so let's get rid of them

df.drop(['Age', 'Arena', 'Attend.', 'Attend./G'], axis = 1, inplace = True)
```

```python
df.columns
```

```python
# the table duplicated column names for several stats (team & oppenent) so let's fix them

df.rename(columns={'eFG%.1': 'OeFG%', 'TOV%.1': 'OTOV%', 'FT/FGA.1': 'OFT/FGA'}, inplace=True)
df.columns
```

```python
# exporting the stats for later use

df.to_csv("./2020_team_stats.csv")
```

```python
df.corr()
```

```python
# looking at all of the numeric features (above) to see if there is a linear correlation
# between the feature and team wins

fig = plt.figure(figsize=(15,50))

for i, col in enumerate(df[1:]):
    fig.add_subplot(13, 3, 1 + i)
    plt.scatter(df[col], df['W'])
    plt.xlabel(col)
    plt.ylabel('Wins')
fig.tight_layout()
fig.show()
```

## Conclusions

Looking at this data it looks like the strongest positive correlations to wins in the NBA involve higher offensive shooting percentages, and negative correlation to wins factor in when the opposing team shoots well. As you would expect, Net Rating is very linearly correlated with number of wins as the "better" teams are much more likely to win games. 

An interesting outlier can be found in the "Pace" scatter plot where the Milwaukee Bucks not only lead the league in pace, but also have the most wins. Another interesting outlier can be found in the Offensive Rating plot where we can see that despite their high offensive rating, the Boston Celtics have underperformed their number of wins.

Most importantly for modeling we clearly visually see the importance of the 4 (really 8 factoring in two-way play) Factors we will be using for our modeling: overall scoring efficiency, free throw shooting, turnover percentage, and rebounding. 

Additional EDA steps will include be further research into incorporating rest days into the final model as star players taking games off strongly affects the outcome of the game.
