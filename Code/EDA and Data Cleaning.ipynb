```python
# since our box score method reads in duplicate sets for each game (one for each team), let's get rid of duplicate 
# and NaN values

data.dropna().drop_duplicates()
print(data.isnull().sum(axis=0).sort_values(ascending=False))
```

```python
# removing unneccesary columns

data.drop(['Unnamed: 0', 'date', 'location', 'losing_abbr', 'losing_name', 'winner', 'winning_abbr',
          'losing_abbr', 'winning_name', 'losing_name'], axis = 1, inplace = True)
```
