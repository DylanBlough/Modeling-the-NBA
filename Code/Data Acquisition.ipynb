The data has been output into the Datasets file with the the format /datasets/{}_boxscores.csv.format(season) so we need to read them back in and combine them.

```python
def read_data(season):
    df = pd.read_csv(f'../datasets/{season}_boxscores.csv') #reads in the data
    
    return df
```

```python
# testing function on one season

read_data(2011)
```

```python
# list of seasons we need

seasons = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
```

```python
# reading in one

dfs = dfs.append(read_data(2011))
```

```python
# adds every season

dfs = pd.DataFrame()
for season in seasons:
    dfs = dfs.append(read_data(season))
```
